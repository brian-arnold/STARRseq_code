```python
species = "Apur"
min_biorep_support = 3 # minimum number of bioreps to support a peak
qval_threshold = 0 # 1.3 corresponds to pvalue of 0.05, 1 corresponds to pvalue of 0.1
macs_folddiff_threshold = 2 # default 2

base_dir = "/Genomics/kocherlab/bjarnold/STARRseq/data"
output_data_deepstarr_dir = f"{base_dir}/deepSTARR_input/min_biorep_support_{min_biorep_support}/qval_{qval_threshold}_folddiff_{macs_folddiff_threshold}/{species}"


random_file = f'{output_data_deepstarr_dir}/random_seqs.fa_predictions_DeepSTARR.model.txt'
dedup_file = f'{output_data_deepstarr_dir}/MACS2_peaks_dedup.fa_predictions_DeepSTARR.model.txt'
raw_file = f'{output_data_deepstarr_dir}/MACS2_peaks_raw.fa_predictions_DeepSTARR.model.txt'

random_df = pd.read_csv(random_file, sep='\t')
dedup_df = pd.read_csv(dedup_file, sep='\t')
raw_df = pd.read_csv(raw_file, sep='\t')
```

```python
def greter_than_fold_diff(df, fd):
    df = df[(df['Predictions_dev'] >= fd) | (df['Predictions_hk'] >= fd)]
    return len(df)


print(f"Fraction of random peaks with greater than {macs_folddiff_threshold} fold difference: ")
print(greter_than_fold_diff(random_df, macs_folddiff_threshold)/len(random_df), greter_than_fold_diff(random_df, macs_folddiff_threshold), len(random_df))

print(f"Fraction of dedup peaks with greater than {macs_folddiff_threshold} fold difference: ")
print(greter_than_fold_diff(dedup_df, macs_folddiff_threshold)/len(dedup_df), greter_than_fold_diff(dedup_df, macs_folddiff_threshold), len(dedup_df))

print(f"Fraction of raw peaks with greater than {macs_folddiff_threshold} fold difference: ")
print(greter_than_fold_diff(raw_df, macs_folddiff_threshold)/len(raw_df), greter_than_fold_diff(raw_df, macs_folddiff_threshold), len(raw_df))
```
